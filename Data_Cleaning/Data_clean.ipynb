# Describe Lattitude and Longtitude and Impute values by  mean

```python
data.loc[:,['Lattitude','Longtitude']].describe() #getting information about columns
```

```python
#Now replace missing values with mean values
data['Lattitude'].mean()
```

```python
data.loc[:,'Lattitude'].fillna(data['Lattitude'].mean(),inplace= True) #replace missing values with mean value
```

```python
data.loc[:,'Longtitude'].fillna(data['Longtitude'].mean(),inplace= True) # fillna() = fill all, where value is nan
```

```python
round(data.isnull().sum(axis = 0).sort_values(ascending = False)/len(data)*100,2)
```

```python
data.loc[:,['Bathroom','Car']].describe()
```

```python
data.groupby('Car').Car.count().sort_values(ascending = False) #count car park space we have highest 2 car parking space in 5606 data
```

```python
data.loc[:,'Car'].fillna(2,inplace= True ) #place 2 in missing value of car
```

```python
data.groupby('Bathroom').Bathroom.count().sort_values(ascending = False)
```

```python
data.loc[:,'Bathroom'].fillna(1,inplace= True )
```

```python
round(data.isnull().sum(axis = 0).sort_values(ascending = False)/len(data)*100,2) #clean data perfectly
```
